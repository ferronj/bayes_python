```python
def sum_income(income_dict: dict) -> float:
    income_values = []
    for value in income_dict.values():
        if not isinstance(value, str):
            income_values.append(value)
    return sum(income_values)


income = pdx_wine['service income']

for each in income:
    revenue = sum_income(income[each]['revenue'])
    expenses = sum_income(income[each]['expenses'])
    profit_per_event = revenue - expenses
    if 'fte_hours' in income[each].keys():
        fte_hours = sum_income(income[each]['fte_hours'])
        print(f'{each}: \n revenue: ${revenue} \n expenses: ${expenses} \n extimated profits per event type: ${profit_per_event} \n employee hours: {fte_hours}')
    else:
        print(f'{each}: \n revenue: ${revenue} \n expenses: ${expenses} \n extimated profits per event type: ${profit_per_event}')
```
